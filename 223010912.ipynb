```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("/content/dataset.csv")
print(data)
```

```python
data.info()
```

```python
from sklearn.preprocessing import LabelEncoder

lbl= LabelEncoder()
data['diagnosis'] = lbl.fit_transform(data['diagnosis'])
data['diagnosis']
```

```python
data.drop(['id'],axis =1, inplace = True)
data.info()
```

```python
from sklearn.preprocessing import MinMaxScaler

target_col = 'diagnosis'

X = data.drop(target_col, axis=1)
Y = data[target_col]

cols = X.columns

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

X_dframe = pd.DataFrame(X_scaled, columns=cols, index=X.index)
X_dframe.tail()
```

```python
from sklearn.model_selection import train_test_split

Xtrain, Xtest, Ytrain, Ytest = train_test_split(X_dframe, Y, test_size = 0.2, random_state=42)
```

```python
Xtrain.shape
```

```python
Ytrain.shape
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(Xtrain)

Xtrain['cluster'] = kmeans.labels_
Xtest['cluster'] = kmeans.predict(Xtest)

Xtrain
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)

knn.fit(Xtrain, Ytrain)

Ypred = knn.predict(Xtest)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
accuracy = accuracy_score(Ytest, Ypred)

precision = precision_score(Ytest, Ypred)

recall = recall_score(Ytest, Ypred)

f1 = f1_score(Ytest, Ypred)

print("Accuracy :", accuracy)
print("Precision:", precision)
print("Recall   :", recall)
print("F1-score :", f1)
```
